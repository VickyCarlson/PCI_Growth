```python
# State Data
HOST = "https://api.census.gov/data"
year = "2005"
dataset = "acs/acs1"
base_url = "/".join([HOST, year, dataset])
predicates = {}
get_vars = ["NAME", "B15002_001E", "B15002_002E","B15002_003E","B15002_004E",
            "B15002_005E","B15002_006E","B15002_007E","B15002_008E","B15002_009E",
            "B15002_010E","B15002_011E","B15002_012E","B15002_013E","B15002_014E",
            "B15002_015E","B15002_016E","B15002_017E","B15002_018E","B15002_019E",
            "B15002_020E","B15002_021E","B15002_022E","B15002_023E","B15002_024E",
            "B15002_025E", "B15002_026E","B15002_027E","B15002_028E","B15002_029E",
            "B15002_030E","B15002_031E","B15002_032E","B15002_033E","B15002_034E",
            "B15002_035E"]
predicates["get"] = ",".join(get_vars)
predicates["for"] = "state:*"
r = requests.get(base_url, params=predicates)
```

```python
print(r.text)
```

```python
# Apply labels to column names
col_names = [
 'State',
 'Population',
 'Population-M',
 'No schooling completed-M',
 'Nursery school to 4th Grade-M',
 '5th and 6th-M',
 '7th and 8th-M',
 '9th grade-M',
 '10th grade-M',
 '11th grade-M',
 '12th grade, no diploma-M',
 'High school_includes equivalency-M',
 'Some college, less than 1 year-M',
 'Some college, 1 or more years, no degree-M',
 "Associates_degree-M",
 "Bachelors_degree-M",
 "Masters degree-M",
 'Professional school degree-M',
 'Doctorate degree-M',
 'Population-F',
 'No schooling completed-F',
 'Nursery school to 4th Grade-F',
 '5th and 6th-F',
 '7th and 8th-F',
 '9th grade-F',
 '10th grade-F',
 '11th grade-F',
 '12th grade, no diploma-F',
 'High school_includes equivalency-F',
 'Some college, less than 1 year-F',
 'Some college, 1 or more years, no degree-F',
 "Associates_degree-F",
 "Bachelors_degree-F",
 "Masters degree-F",
 'Professional school degree-F',
 'Doctorate degree-F',
'State_No'
]
```

```python
# Create Dataframe
df = pd.DataFrame(columns=col_names, data=r.json()[1:])
df
```

```python
# Insert year for dataset
df.insert(1, 'YEAR', year)
```

```python
# Drop Puerto Rico, District of Columbia, State_No
df = df[df.State_No != '72']
df = df[df.State_No != '11']
df
```

```python
# Drop state number
df.drop('State_No', axis=1, inplace=True)
df
```

```python
# View data types
df.info()
```

```python
# Change data types
df['State'] = df['State'].astype('string')
df['YEAR'] = df['YEAR'].astype(int)
df['Population'] = df['Population'].astype(int)
df['Population-M'] = df['Population-M'].astype(int)
df['No schooling completed-M'] = df['No schooling completed-M'].astype(int)
df['Nursery school to 4th Grade-M'] = df['Nursery school to 4th Grade-M'].astype(int)
df['5th and 6th-M'] = df['5th and 6th-M'].astype(int)
df['7th and 8th-M'] = df['7th and 8th-M'].astype(int)
df['9th grade-M'] = df['9th grade-M'].astype(int)
df['10th grade-M'] = df['10th grade-M'].astype(int)
df['11th grade-M'] = df['11th grade-M'].astype(int)
df['12th grade, no diploma-M'] = df['12th grade, no diploma-M'].astype(int)
df['High school_includes equivalency-M'] = df['High school_includes equivalency-M'].astype(int)
df['Some college, less than 1 year-M'] = df['Some college, less than 1 year-M'].astype(int)
df['Some college, 1 or more years, no degree-M'] = df['Some college, 1 or more years, no degree-M'].astype(int)
df["Associates_degree-M"] = df["Associates_degree-M"].astype(int)
df["Bachelors_degree-M"] = df["Bachelors_degree-M"].astype(int)
df["Masters degree-M"] = df["Masters degree-M"].astype(int)
df['Professional school degree-M'] = df['Professional school degree-M'].astype(int)
df['Doctorate degree-M'] = df['Doctorate degree-M'].astype(int)
df['Population-F'] = df['Population-F'].astype(int)
df['No schooling completed-F'] = df['No schooling completed-F'].astype(int)
df['Nursery school to 4th Grade-F'] = df['Nursery school to 4th Grade-F'].astype(int)
df['5th and 6th-F'] = df['5th and 6th-F'].astype(int)
df['7th and 8th-F'] = df['7th and 8th-F'].astype(int)
df['9th grade-F'] = df['9th grade-F'].astype(int)
df['10th grade-F'] = df['10th grade-F'].astype(int)
df['11th grade-F'] = df['11th grade-F'].astype(int)
df['12th grade, no diploma-F'] = df['12th grade, no diploma-F'].astype(int)
df['High school_includes equivalency-F'] = df['High school_includes equivalency-F'].astype(int)
df['Some college, less than 1 year-F'] = df['Some college, less than 1 year-F'].astype(int)
df['Some college, 1 or more years, no degree-F'] = df['Some college, 1 or more years, no degree-F'].astype(int)
df["Associates_degree-F"] = df["Associates_degree-F"].astype(int)
df["Bachelors_degree-F"] = df["Bachelors_degree-F"].astype(int)
df["Masters degree-F"] = df["Masters degree-F"].astype(int)
df['Professional school degree-F'] = df['Professional school degree-F'].astype(int)
df['Doctorate degree-F'] = df['Doctorate degree-F'].astype(int)
df.info()
```

```python
# Combine columns, drop unnecessary columns
df['Less_than_9th_ grade'] = df['No schooling completed-M'] + df['No schooling completed-F'] \
+ df['Nursery school to 4th Grade-M'] + df['Nursery school to 4th Grade-F'] \
+ df['5th and 6th-M'] + df['5th and 6th-F'] + df['7th and 8th-M'] + df['7th and 8th-F']
df.drop(df.iloc[:, 3:8], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['9th_to_12th_grade_no_diploma'] = df['9th grade-M'] + df['10th grade-M'] + df['11th grade-M'] \
+ df['12th grade, no diploma-M'] + df['9th grade-F'] + df['10th grade-F'] + df['11th grade-F'] \
+ df['12th grade, no diploma-F']
df.drop(df.iloc[:, 3:7], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['High_school_graduate_(includes_equivalency)'] = df['High school_includes equivalency-M'] + df['High school_includes equivalency-F']
df.drop(df.iloc[:, 3:4], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['Some_college,_no_degree'] = df['Some college, less than 1 year-M'] + df['Some college, less than 1 year-F'] \
+ df['Some college, 1 or more years, no degree-M'] + df['Some college, 1 or more years, no degree-F']
df.drop(df.iloc[:, 3:5], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['Associates_degree'] = df['Associates_degree-M'] + df['Associates_degree-F'] 
df.drop(df.iloc[:, 3:4], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['Bachelors_degree'] = df['Bachelors_degree-M'] + df['Bachelors_degree-F'] 
df.drop(df.iloc[:, 3:4], inplace = True, axis =1)
df
```

```python
# Combine columns, drop unnecessary columns
df['Graduate_or_professional_degree'] = df['Masters degree-M'] + df['Masters degree-F'] \
+ df['Professional school degree-M'] + df['Professional school degree-F'] \
+ df['Doctorate degree-M'] + df['Doctorate degree-F']
df.drop(df.iloc[:, 3:6], inplace = True, axis =1)
df
```
